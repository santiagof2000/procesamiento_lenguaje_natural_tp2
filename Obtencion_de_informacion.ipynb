```python
import requests

def scrap(url):
    partes = url.split("/")

    anio = partes[-1]

    # Realizamos una solicitud HTTP GET a la URL
    response = requests.get(url, verify=False)

    # Parseamos el contenido de la pÃ¡gina web utilizando BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Inicializamos una cadena para almacenar el texto
    texto_del_articulo = ""

    try:
        parrafos = soup.find_all('p')

        # Concatenamos los pÃ¡rrafos para obtener el texto completo
        for parrafo in parrafos:
            texto_del_articulo += parrafo.text

    except Exception as e:
        print('error')

    # Retornamos la lista de diccionarios
    return texto_del_articulo
```

```python
import re

def limpieza(text):
    text = re.sub(r'\u200b', '', text)
    text = re.sub(r'\[[^\]]*?\]', '', text)
    return text
```

```python
!pip install reportlab
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Paragraph
from reportlab.lib.styles import getSampleStyleSheet

def string_to_pdf(input_string, filename):
    doc = SimpleDocTemplate(filename, pagesize=letter)
    styles = getSampleStyleSheet()
    story = []
    for line in input_string.split('\n'):
        story.append(Paragraph(line, styles["BodyText"]))
    doc.build(story)

```
